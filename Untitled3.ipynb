```python

import numpy as np
import matplotlib.pyplot as plt

# 상자의 크기
box_width = 10.0
box_height = 10.0

# 분자의 초기 상태 설정
num_particles = 100
positions = np.random.rand(num_particles, 2) * [box_width, box_height]  # 초기 위치
velocities = np.random.randn(num_particles, 2)  # 초기 속도 (정규분포)

# 시뮬레이션 매개변수
num_steps = 1000  # 시뮬레이션 단계 수
dt = 0.01  # 시간 간격

# 시뮬레이션 루프
for step in range(num_steps):
    # 벽에 닿으면 반사 처리
    out_of_bounds = np.abs(positions) >= [box_width, box_height]
    velocities[out_of_bounds] *= -1

    # 위치 업데이트
    positions += velocities * dt

# 결과 출력
plt.figure(figsize=(6, 6))
plt.scatter(positions[:, 0], positions[:, 1], s=10)
plt.xlim(0, box_width)
plt.ylim(0, box_height)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Monte Carlo Simulation of Gas Molecules')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 초기 설정
num_particles = 100  # 기체 분자 수
num_steps = 1000  # 시뮬레이션 단계 수
box_length = 10.0  # 상자의 길이

# 초기 위치 설정 (랜덤으로 상자 내부에 배치)
positions = np.random.rand(num_particles) * box_length

# 초기 속도 설정 (정규분포를 따르는 랜덤 속도)
velocities = np.random.randn(num_particles)

# 시뮬레이션 루프
for step in range(num_steps):
    # 위치 업데이트
    positions += velocities

    # 상자 경계 조건 처리 (반사)
    positions[positions < 0] *= -1
    positions[positions > box_length] = 2 * box_length - positions[positions > box_length]

# 결과 시각화
plt.figure(figsize=(8, 4))
plt.hist(positions, bins=50, density=True, alpha=0.7)
plt.title('Monte Carlo Simulation of Gas Molecules in 1D')
plt.xlabel('Position')
plt.ylabel('Probability Density')
plt.xlim(0, box_length)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 상수 정의
kB = 1.38e-23  # 볼츠만 상수 (J/K)
N_A = 6.022e23  # 아보가드로 수 (1/mol)
R = kB * N_A  # 기체 상수 (J/(mol*K))

# 시뮬레이션 매개변수
num_particles = 1000  # 기체 분자 수
box_volume = 1.0  # 상자의 부피 (m^3)
temperature = 300.0  # 온도 (K)
num_steps = 10000  # 시뮬레이션 단계 수

# 초기 설정
positions = np.random.rand(num_particles, 3) * np.sqrt(box_volume)  # 초기 위치 (균일하게 분포)
velocities = np.random.randn(num_particles, 3) * np.sqrt(kB * temperature / R)  # 초기 속도 (Maxwell-Boltzmann 분포)

# 시뮬레이션 루프
pressure_accumulator = 0.0
for step in range(num_steps):
    # 위치 업데이트 (주어진 시간 간격에서 속도에 따라 위치를 업데이트)
    positions += velocities * (1.0 / num_steps)

    # 경계 조건 (상자 내부에 위치하도록)
    positions = np.mod(positions, np.sqrt(box_volume))

    # 압력 계산 (각 축 방향에서의 입자 운동에 따른 운동량 변화 계산)
    momentum_change = np.sum(np.abs(velocities))
    pressure_accumulator += momentum_change / (3 * box_volume)

# 평균 압력 계산
average_pressure = pressure_accumulator / num_steps

# 이론적 상태 방정식에서의 예상 압력 계산
expected_pressure = (N_A * temperature) / box_volume

# 결과 출력
print(f"Average pressure from simulation: {average_pressure:.4f} Pa")
print(f"Expected pressure from ideal gas law: {expected_pressure:.4f} Pa")

# 시뮬레이션 결과와 이론적 예상값 비교
plt.figure(figsize=(8, 6))
plt.hist(positions[:, 0], bins=50, density=True, alpha=0.7, label='Histogram of Positions')
plt.title('Monte Carlo Simulation of Gas Molecules')
plt.xlabel('Position')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 상수 정의
kB = 1.38e-23  # 볼츠만 상수 (J/K)
N_A = 6.022e23  # 아보가드로 수 (1/mol)
R = kB * N_A  # 기체 상수 (J/(mol*K))

# 시뮬레이션 매개변수
num_particles = 1000  # 기체 분자 수
box_volume = 1.0  # 상자의 부피 (m^3)
temperature = 300.0  # 온도 (K)
num_steps = 1000  # 시뮬레이션 단계 수

# 초기 설정
positions = np.random.rand(num_particles, 3) * np.sqrt(box_volume)  # 초기 위치 (균일하게 분포)
velocities = np.random.randn(num_particles, 3) * np.sqrt(kB * temperature / R)  # 초기 속도 (Maxwell-Boltzmann 분포)

# 시뮬레이션 루프
trajectory = np.zeros((num_steps, num_particles, 3))
for step in range(num_steps):
    # 위치 업데이트 (주어진 시간 간격에서 속도에 따라 위치를 업데이트)
    positions += velocities * (1.0 / num_steps)
    trajectory[step] = positions

    # 경계 조건 (상자 내부에 위치하도록)
    positions = np.mod(positions, np.sqrt(box_volume))

# 결과 시각화 (분자의 운동 패턴 시각화)
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

for particle in range(num_particles):
    ax.plot(trajectory[:, particle, 0], trajectory[:, particle, 1], trajectory[:, particle, 2])

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Trajectory of Gas Molecules in 3D Box')
plt.show()

```
